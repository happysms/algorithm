# DFS

- DFS는 깊이 우선 탐색이라고도 부르며 그래에서 깊은 부분을 우선적으로 탐색하는 알고리즘.
- DFS는 스택 자료구조(혹은 재귀 함수)를 이용하며, 구체적인 동작 과정은 다음과 같다. 
    1. 탐색 시작 노드를 스택에 삽입하고 방문 처리를 한다. 
    2. 스택의 최상단 노드에 방문하지 않은 인접한 노드가 하나라도 있으면 그 노드를 스택에 넣고 방문 처리를 한다. 방문하지 않은 인접 노드가 없으면 스택에서 최상단 노드를 꺼낸다. 
    3. 더 이상 2번의 과정을 수행할 수 없을 때까지 반복한다. 

```python
# DFS 메서드 정의
def dfs(graph , v, visited):
    #현재 노드를 방문 처리
    visited[v] = True
    print(v, end = ' ')
    #현재 노드와 연결된 다른 노드를 재귀적으로 방문
    for i in graph[v]:
        if not visited[i]:
            dfs(graph , i , visited)
```

```python
#2667번 단지번호붙히기

dx = [0 , 1 , 0 , -1]
dy = [-1, 0 , 1, 0]

n = int(input())
arr = []
visited = []
result = []
num = 0

for i in range(n):
    arr.append(list(map(int, input())))
    visited.append([False for i in range(n)])
    
def dfs(x , y):
    visited[x][y] = True
    global num
    num += 1
    for i in range(4):
        new_x = x + dx[i]
        new_y = y + dy[i]
    
        if new_x >= 0 and new_x < n and new_y >= 0 and new_y < n and (visited[new_x][new_y] == False) and arr[new_x][new_y] == 1: 
            dfs(new_x , new_y)

for i in range(n):
    for j in range(n):
        if arr[i][j] == 1 and visited[i][j] == False:
            dfs(i, j)
            result.append(num)
            num = 0

print(len(result))
result.sort()

for i in result:
    print(i)
```

```python
# 11724번 연결 요소의 개수 DFS풀이
 
n , m = map(int, input().split())

data = [[] for i in range(n + 1)]
visited = [False] * (n + 1)

for i in range(m):
    a, b = map(int, input().split())
    data[a].append(b)
    
def dfs(start):
    global cnt
    cnt += 1
    visited[start] = True
    for i in data[start]:
        if visited[i] == False:
            dfs(i)
        
cnt = 0
result = []

for i in range(1 , n + 1):
    if visited[i] != True:
        dfs(i)
        result.append(cnt)
        cnt = 0

print(len(result))
```

```python
# 11724번 BFS풀이
import sys
from collections import deque

n , m = map(int, input().split())
data = [[] for i in range(n + 1)]
visited = [0] * (n + 1)


for i in range(m):
    a, b = map(int, input().split())
    data[a].append(b)
    data[b].append(a)

def bfs():
    q = deque()
    result = 0
    for i in range(1, n + 1):
        if visited[i] == 0:
            visited[i] = 1
            q.append(i)
            result += 1
            while q:
                now = q.popleft()
                for j in data[now]:
                    if visited[j] == 0:
                        visited[j] = 1
                        q.append(j)
    print(result)
    
bfs()
    
```
