```python
# 다익스트라 알고리즘

inf = 1000

# W = [[0 , 2, 4, 1, i, i],
#     [i, 0 ,3, i, i, 1],
#     [i, i, 0 , inf, 1 , 5],
#     [inf, inf, 2, 0 , 6 , inf],
#     [inf, inf, inf, inf, 0 , 2],
#      [inf, inf, inf, inf, inf, 0]]

n ,m = map(int, input("노드의 개수, 간선의 개수 : ").split())

W = [[inf] * n for i in range(n)]

for i in range(m):
    a , b, weight = map(int, input("출발 노드, 도착 노드, 가중치 : ").split())
    W[a - 1][b - 1] = weight
    
f = set()
touch = n * [0]
length = n * [0]
save_length = n * [0]
vnear = 0

for i in range(1, n):
  touch[i] = 0
  length[i] = W[0][i]

for _ in range(n - 1):
  min = 1000
  for i in range(1, n):
    if 0 <= length[i] < min:
      min = length[i]
      vnear = i
  
  e = (touch[vnear] , vnear)
  f.add(e)

  for i in range(1, n):
    if length[vnear] + W[vnear][i] < length[i]:
      length[i] = length[vnear] + W[vnear][i]
      touch[i] = vnear
      save_length[i] = length[vnear] + W[vnear][i]

  save_length[vnear] = length[vnear]
  length[vnear] = -1

print(f)
print(save_length)
print(touch)
```

```python
# 허프만 코드 
import heapq
from collections import defaultdict

def encode(frequency):
    heap = [[weight, [symbol, '']] for symbol, weight in frequency.items()]
    print(heap)
    heapq.heapify(heap)
    while len(heap) > 1:
        lo = heapq.heappop(heap)
        hi = heapq.heappop(heap)
        for pair in lo[1:]:
            pair[1] = '0' + pair[1]
        for pair in hi[1:]:
            pair[1] = '1' + pair[1]

        heapq.heappush(heap, [lo[0] + hi[0]] + lo[1:] + hi[1:])
    return sorted(heapq.heappop(heap)[1:], key=lambda p: (len(p[-1]), p))

data = "The frog at the bottom of the well drifts off into the great ocean"
frequency = defaultdict(int)

for symbol in data:
    frequency[symbol] += 1

huff = encode(frequency)
print ("Symbol".ljust(10) + "Weight".ljust(10) + "Huffman Code")
for p in huff:
    print (p[0].ljust(10) + str(frequency[p[0]]).ljust(10) + p[1])
```

```python
#4-Queens 문제의 상태공간트리(되추적)
result = []

def promising(i, col): 
    switch = True
    k = 0  # 현재 행보다 작은 행들의 정보를 탐색 
    while k < i and switch:
        if col[i] == col[k] or abs(col[i] - col[k]) == i - k: # 같은 column인지 확인 & 같은 대각에 있는지 확인
            switch = False
        k += 1 # switch가 False가 아니기에 다음 행을 탐색한다. 
    
    return switch # 최종적으로 promising 인지 아닌지 반환하는 것

def queens(n, i, col):
    if promising(i, col):
        if i == n - 1: # 끝까지 탐색했다면
            print(col)
            temp = col[:]
            result.append(temp)
        else:
            for j in range(0, n):
                col[i + 1] = j # 같은 i + 1 열에 있는 것은 모두 j
                queens(n ,i + 1, col)
                


n = int(input("Length : "))
col = n * [0]
queens(n, -1 , col)
print("해의 개수 : ", len(result))
print("3번째 해 : " , result[2])
```

```python
# 배낭 채우기 최고우선탐색 

from queue import PriorityQueue

n = 4  # 물건 개수
W = 16 # 채울 무게
p = [40, 30, 50 , 10] # 가치
w = [2, 5, 10, 5] # 각 물건의 무게
include = [0] * n
maxprofit = 0
bestset = n * [0]

class Node:
    def __init__(self, level, profit, weight, include):
        self.level = level
        self.profit = profit
        self.weight = weight
        self.bound = 0
        self.include = include[:]

    def print(self):
        print(self.level,self.profit,self.weight,self.bound,self.include)
        
    def __lt__(self, other):
        return self.bound < other.bound
    
    def __le__(self, other):
        return self.bound <= other.bound
    
    def __gt__(self, other):
        return self.bound > other.bound
    
    def __ge__(self, other):
        return self.bound >= other.bound
    
    def __eq__(self, other):
        return self.bound == other.bound
    

def knapsack():
    global maxprofit
    global bestset
    temp = n * [0]
    P = PriorityQueue()
    v = Node(-1,0,0,temp)

    v.bound = compbound(v)
    P.put((v.bound,v))
    
    while(not P.empty()):
        v = P.get()[1] 
        print(v.include) 
        
        if v.bound < maxprofit and v.level != n - 1:
            level = v.level + 1
            weight = v.weight + w[level]
            profit = v.profit + p[level]
            temp = v.include[:]
            u = Node(level, profit, weight, temp)

            if u.weight <= W and u.profit > maxprofit:
                maxprofit = u.profit
                bestset = u.include[:]
                
            u.bound = compbound(u)
            u.include[u.level] = 1
            
            if(-u.bound > maxprofit):
                P.put((u.bound,u))
                
            u = Node(level, v.profit, v.weight, temp)
            u.include[u.level] = 0
            
            u.bound = compbound(u)
            if(-u.bound > maxprofit):
                P.put((u.bound,u))
            if u.weight <= W and u.profit > maxprofit:
                maxprofit = u.profit
                bestset = u.include[:]
        
    
def compbound(u):     
    j = u.level + 1
    bound = u.profit
    totweight = u.weight
    
    while j < n and totweight + w[j] <= W:
        totweight = totweight + w[j]
        bound = bound + p[j]
        j += 1
    
    k = j
    if k < n:
        bound = bound + (W - totweight) * (p[k] / w[k])
    
    return -bound


knapsack()
print("maxprofit = ",maxprofit)
```

```python
# 부분집합의 합

def promising1(i , weight, total):
    if ((weight + total) >= W) and (weight == W or (weight + S[i + 1]) <= W):
        return True
    else:
        return False

def s_s(i , weight, total, include):
    if (promising1(i, weight, total)):
        if weight == W:
            print(include)
        else:
            include[i + 1] = 1
            s_s(i + 1 , weight + S[i + 1], total - S[i + 1], include)
            include[i + 1] = 0
            s_s(i + 1, weight, total - S[i + 1], include)

n = 4
S = [1,2,4,6]
S.sort()
W = 6
include = n * [0]
total = 0
for k in S:
    total += k

s_s(-1, 0, total, include)
```
